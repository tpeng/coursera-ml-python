```python
import numpy as np

def load_dataset(fname='ex2data1.txt'):
    """
    load a (X,y) tuple while X is a n_samples * n_features numpy array and y is the label.
    """
    X = []
    y = []
    for line in open(fname):
        rows = [float(x) for x in line.split(',')]
        X.append(rows[0:2])
        y.append(rows[-1])
    return np.array(X, float32), np.array(y, int)

X, y = load_dataset()
```

```python
import matplotlib.pyplot as plt

positive = X[y==1]
negative = X[y==0]

x1 = positive[:, 0]
y1 = positive[:, 1]
plt.plot(x1, y1, 'k+')

x2 = negative[:, 0]
y2 = negative[:, 1]
plt.plot(x2, y2, 'ko')

xlabel('Exam 1 score')
ylabel('Exam 2 score')

plt.show()
```

## Cost function的推导

- sigmoid函数 $$ h(\theta) = \frac{1}{1+ e^{-w*x}} $$

- $$ p(y = 1|x;\theta) = h_{\theta}(x) $$ 
- $$ p(y = 0|x;\theta) = 1- h_{\theta}(x) $$ 

可以写成更加紧凑的形式:

$$ p(y| x; \theta) = h_{\theta}(x)^y * (1 - h_{\theta}(x) ^ {1 - y} $$

(因为y=1时, 后者为1, y=0时, 前者为1)

假设是独立的, 所以似然函数定义如下:

$$ L(\theta) = \prod p(y^i | \theta; x^i) $$

而对数似然函数是:

$$ log(L) = \sum_{i=1}^{m} y^{i} log(h(\theta)) + (1 - y^i) log (1 - h(\theta)) $$
->
$$ \sum_{i=1}^{m} y^i \frac{h(\theta)}{1 - h(\theta)} - log (1 - h(\theta)) $$


因为 $$ log \frac{h(\theta)}{1-h(\theta)} = \theta *x $$

所以上式写成:

$$ \sum_{i=1}^{m} (y^i * \theta * x - log (1 + e^{\theta*x}) $$

对$$$\theta$$$求偏导有:

$$ y^{i} x^{i} - \frac{e^{\theta*x} * x^{i}}{1+e^{\theta*x}} $$

而 $$ h(\theta) = \frac{1}{1+ e^{-w*x}} = \frac{e^{\theta*x}}{1+e^{\theta*x}} $$ 所以偏导可以写成:

$$  (y^{i} - h_{\theta}(x)) * x^{i} $$

因为linear regression中用gradient descent求极小值, 而此处的log似然函数要求极大值, 所以把cost function至反, 从而求极小值.
